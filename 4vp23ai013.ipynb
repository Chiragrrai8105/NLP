```python
!pip install nltk
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwards
from nltk.stem import PorterStemmer
import re

nltk.download('punkt_tab')
nltk.download('stopwards')

def preprocess_text(text):
    tokens=word_tokenize(text)
    print("Tokenized words:",tokens)
    filtered_tokens=[word for word in tokens if word.isalnum()] 
    print("Filtered Tokens:",filtered_tokens)
    validated_tokens=[word for word in filtered_tokens if re.match("^[A-Za-z]+$",word)]
    print("Validated Tokens:",validated_tokens)
    stop_words=set(stopwards.words('english'))
    tokens_without_stopwards=[word for word in filtered_tokens if word.lower() not in stop_words]
    print("Tokens after stop word removal:",tokens_without_stopwards)
    stemmer=PorterStemmer()
    stemmed_tokens=[stemmer.stem(word) for word in tokens_without_stopwards]
    print("Stemmed tokens:",stemmed_tokens)
    return stemmed_tokens

text="Natural Language Processing(NLP) is a fascinating field of Artificial Integillegence!"
preprocessed_text=preprocess_text(text)
    

    
```
